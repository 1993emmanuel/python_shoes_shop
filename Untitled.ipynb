```python
import math
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# importamos nuestro csv para poder manipular los datasets
df = pd.read_csv( 'shoes_dataset.csv' )
```

```python
# para mostrar los 5 registros y saber la informacion de nuestro CSV
df.head()
```

```python
# CONVENRTIRLO A DATE TIME
# con la funcion .info() nos muestran los tipo de datos de cada una de nuestras columnas
df.info()
```

```python
#CONVERTIMOS DATE A DATETIME
df['Date'] = pd.to_datetime(df['Date'])
# ejecutamos de nuevo info para saber si cambio el valor
df.info()
```

```python
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
# separamos creando nuevas columnas con los valores a√±o dia y mes ya cada uno de manera
# numerica.
# el acomodo es year mont day pero podemos acomodarlo como mas facil se nos haga
df.info()
```

```python
df.head()
```

```python
# sale price es un string y lo necesitamos de tipo numerico para realizar los calculos
# tiene un signo de $ y lo tenemos que quitar
# tambien quitamos el de Unite price
df['SalePrice'] = df['SalePrice'].apply( lambda x: float(x[2:]) )
df['UnitPrice'] = df['UnitPrice'].apply( lambda x : float(x[2:]) )
```

```python
# REALIZAREMOS DIBUJADOS CON SEABORN
# SEPARAMOS NUESTRA VARIABLES CATEGORICAS Y LAS NUMERICAS
categorical_variables = ['Country', 'ProductID','Shop','Gender','Size (US)', 'Discount','Year','Month']
numerical_variables = ['UnitPrice','SalePrice']
categoria_country = ['Country']
categoria_shop = ['Shop']
categoria_Gender = ['Gender']
```

```python
# este grafico solo sera para el country
for cat_country in categoria_country:
    frecuencias = df[cat_country].value_counts()
    df_frecuencia = pd.DataFrame({ cat_country: frecuencias.index.tolist(), 'cantidad' : frecuencias.tolist() })
    sns.barplot( x=cat_country, y='cantidad', data=df_frecuencia )
    plt.show()
```

```python
# ahora mostraremos en las tiendas
for cat_shop in categoria_shop:
    frecuencias = df[cat_shop].value_counts()
    df_frecuencia = pd.DataFrame({ cat_shop : frecuencias.index.tolist(), 'cantidad' : frecuencias.tolist() })
    sns.barplot( x=cat_shop, y='cantidad', data= df_frecuencia )
    plt.show()
```

```python
# para realizar el graficado haremos un loop por todas nuestras varibles categoricas
# explicando el for loop
# vamos a recorrer todas las categorical variables
for cat_variable in categorical_variables:
    # generamos una variable frecuency que es igual a nuestro dataframe en la posicion cat variable
    #y haremos que cuente los valores ya sea de los diferentes country que tengamos
    frecuency = df[cat_variable].value_counts()
    # nuestro df frecuendy es igual a un nuevo dataframe que tendra cat_variable y sacaremos la frecuentcia obtenida
    # anteriormente la haremos lista y el valor sera por la cantidad $$$ en el valor de la parte inferior
    df_frecuency = pd.DataFrame({cat_variable: frecuency.index.tolist(), 'quantity' : frecuency.tolist() })
    # ahora creamos el grafico en el plano x sera la cat_varibale y el y la cantidad y el data es nuestro
    # data frame df_frecuency 
    sns.barplot( x=cat_variable, y='quantity', data=df_frecuency )
    # mostramos los graficos
    plt.show()
```

```python
# T score es para muestras 
# y Z score para poblaciones
# agruparemos nuestros datos para 

# en nuestro data frame traeremos todos los que no son 2014(2015 2016) y todos los compradores que sean hombres
# y todos los compradores de estados unidos
# una vez que termine con groupby lo agruparemos por las tres variables size year month despues con size
# nos muestra la cantidad total de los datos con unstack estaran los valores en las columnas
# y con fillna los valores nulos seran remplazados por 0
grouped = df[(df['Year'] != 2014) & (df['Gender'] == 'Male')
             & (df['Country'] == 'United States')].groupby(
             ['Size (US)', 'Year', 'Month']).size()\
             .unstack(level=0).fillna(value=0)

grouped
```

```python
# vamos a calcular la media y el error estandar
# sem nos trae el error estandar
means = []
standar_error = []
for column in grouped.columns:
    means.append( grouped[column].mean() )
    standar_error.append( grouped[column].sem() )
```

```python
means
```

```python
standar_error
```

```python
# meteremos las medias y los estandares en un data frame
d = { 'means' : means, 'estandar_error' : standar_error }
```

```python
#ahora lo vamos a convertir en un data frame de pandas
df_calculation = pd.DataFrame( data= d, index=grouped.columns )
```

```python
# mostramos como queda el dataframe
df_calculation.head(50)
```

```python
# calcularemos el margen de error
df_calculation['error_margin'] = df_calculation['estandar_error'].apply( lambda x: x * 2.07 )
```

```python
df_calculation.head(500)
```

```python
df_calculation['low_margin'] = df_calculation.apply( lambda x : x['means'] - x['error_margin'], axis=1 )
df_calculation['up_margin'] = df_calculation.apply( lambda x : x['means'] - x['error_margin'], axis=1 )
# redondeamos a arriba
df_calculation['math_round_up'] = df_calculation.apply( lambda x : math.ceil(x['up_margin']), axis=1 )
```
